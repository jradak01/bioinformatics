## [BA3C](https://rosalind.info/problems/ba3c/) - Construct the Overlap Graph of a Collection of k-mers

```python
def prefix(a):
  return a[:-1]
def suffix(a):
  return a[1:]
```

```python
def sortDict(D):
  graph = {}
  sorted= list(D.keys())
  sorted.sort()
  for key in sorted:
    graph[key] = D[key]
    graph[key].sort()
  return graph
```

```python
def overlapGraph(kmers):
  D = {}
  for i in range(len(kmers)):
    for j in range(len(kmers)):
      if i != j:
        if prefix(kmers[j]) == suffix(kmers[i]):
          if kmers[i] not in D:
            D[kmers[i]] = [kmers[j]]
          else:
            D[kmers[i]].append(kmers[j])
            D[kmers[i]].sort()
  graph = sortDict(D)
  res = []
  for key, value in graph.items():
    if len(value) > 1:
      res.append(key + ' -> ' + ','.join(value))
    else:
      res.append(key + ' -> ' + value[0])
  return '\n'.join(res)
```

## [BA3D](https://rosalind.info/problems/ba3d/) - Construct the De Bruijn Graph of a String

```python
def deBrujin(k, text):
  D = {}
  for i in range(len(text)-k +1):
    prefix = text[i: i+k-1]
    suffix = text[i+1: i+k]
    if prefix not in D:
      D[prefix] = [suffix]
    else:
      D[prefix].append(suffix)
      D[prefix].sort()
  D = sortDict(D)
  res = []
  for key, value in D.items():
    res.append(key + ' -> ' + ','.join(value))
  return '\n'.join(res)
k = 4
text = 'AAGATTCTCTAC'
print(deBrujin(k, text))
```

```python
# BA3E
def deBrujinFromKmers(kmers):
  D = {}
  for kmer in kmers:
    if prefix(kmer) not in D:
      D[prefix(kmer)] =[suffix(kmer)]
    else:
      D[prefix(kmer)].append(suffix(kmer))
      D[prefix(kmer)].sort()
  D = sortDict(D)
  res = []
  for key, value in D.items():
    res.append(key + ' -> ' + ','.join(value))
  return '\n'.join(res)

kmers = [
'GAGG',
'CAGG',
'GGGG',
'GGGA',
'CAGG',
'AGGG',
'GGAG']
print(deBrujinFromKmers(kmers))
```

## [BA3F](https://rosalind.info/problems/ba3f/) - Find an Eulerian Cycle in a Graph

```python
def getDict(directions):
  D = {} # rijecnik
  for direction in directions: # za svaku uputu
    # razdvoji na ono prije strelice i poslije
    first, second = direction.split(" -> ")
    # ako poslije strelice ima vise uputa, razdvoji
    second = second.split(",")
    D[first] = second # dodaj u rijecnik
  return D
```

```python
def getACycle(graph, possible_starts):
  origin = None
  for start in possible_starts: # za svaki start u mogucim pocetcima
    if start in graph: # ako je start u grafu
      origin = start # postavi izvor na start
      break # prekini petlju
  if origin is None: # ako nema izvora -> iznimka
    raise
  cycle = []
  first = origin # postavi kao prvi izvor
  while True: # dok je istina
    if first in graph: # ako je prvi u grafu
      second = graph[first][0] # dohvati prvu vrijednost iz rijecnika za kljuc
      cycle.append([first, second]) # dodaj u ciklus par prvi-drugi
      if len(graph[first]) == 1: # ako je broj vrijednosti uz kljuc jednaka 1
        graph.pop(first, None) # izbaci kljuc
      else: # inace
        graph[first] = graph[first][1:] # ostavi samo element koji nije prijeden
      if second == origin: # ako je drugi jednak izvoru
        break # prekini petlju
      else:
        first = second # inace postavi prvog na drugog
    else: # inace, ako prvi nije u grafu
      break # prekini
  return cycle # vrati ciklus
```

```python
def start_index(cycle, start):
  s_index = -1 # postavi s_index na neg. broj
  for i in range(0, len(cycle)): # za svaki element u ciklusu
    if cycle[i][0] == start: # ako je drugi element iz ciklusa jednak startu
      s_index = i # postavi index na i
      break # izadi iz petlje
  return s_index # vrati dobiveni index
```

```python
def eulerianCycle(graph, start=None):
  graph = graph.copy() # napravi kopiju grafa
  cycles = []
  possible_starts = list(graph.keys())[:1] # postavi kao moguci pocetak prvi el
  if start != None: # ako je unesen start
    possible_starts = [str(start)] # broj koji cemo stavit za pocetak je uneseni
  while len(graph) > 0: # dok je duljina grafa veca od 0
    # vrati sve moguce krugove od moguceg pocetka
    new_cycle = getACycle(graph, possible_starts) # napravi novi ciklus
    # uzmi svaki prvi element iz novog ciklusa (u parovima)
    # i dodaj u listu mogucih pocetaka
    possible_starts.extend([x[0] for x in new_cycle])
    possible_starts = list(set(possible_starts)) # da se uklone duplikati
    cycles.append(new_cycle) # nadodaj novi ciklus
  if len(cycles) == 1: # ako je duzina ciklusa jednaka 1 (nije bilo grananja)
    # put je jedan niz (u dvostrukim je zagradama radi append u petlji)
    path = cycles[0]
  else: # inace
    path = cycles[0] # uzmi prvi dobiveni put
    for i in range(1, len(cycles)): # za svaki iduci ciklus
      next_ = cycles[i] # next je taj ciklus
      # pronadi index gdje cemo umetnit prvi element
      # ovog ciklusa u pocetni ciklus
      s_index = start_index(path, next_[0][0])
      # umetni element ovog ciklusa u pocetni ciklus
      path = path[:s_index] + next_ + path[s_index:]
  firsts = [str(x[0]) for x in path]
  firsts.append(str(path[-1][1]))
  return '->'.join(firsts)
```

## [BA3G](https://rosalind.info/problems/ba3g/) - Find an Eulerian Path in a Graph

```python
def RebalanceGraphFromEulerianPathToCycle(graph):
  graph = graph.copy() # napravi kopiju grafa
  # broj vrijednosti za svaki kljuc
  outbalance = {first: len(graph[first]) for first in graph.keys()}
  new_list = []
  for v in graph.values(): # za svaku vrijednost u rijecniku
    new_list.extend(v) # nadodaj vrijednost u listu
  inbalance = Counter(new_list) # prebroji pojavljivanja svakog elementa
  # svi jedinstveni elementi koji se pojavljuju i u kljucevima i u vrijednostima
  all_nodes = set(list(outbalance.keys()) + list(inbalance.keys()))
  for node in all_nodes: # za svaki cvor u cvorovima
    # izracunaj razliku broja vrijednosti koje sadrzi
    # kljuc i broja pojavljivanja u vrijednostima
    balance = outbalance.get(node, 0) - inbalance.get(node, 0)
    if balance == 1: # ako je razlika jednaka jedan
      second = node # postavi drugi na taj cvor
    if balance == -1: # ako je razlika -1
      first = node # postavi prvi na cvor
  if first not in graph: # ako prvi nije u grafu u kljucevima
    # dodaj jos jednu vezu prvi - drugi (kljuc - vrijednost)
    graph[first] = [second]
  else: # inace
    # nadodaj drugi (vrijednost) na postojeci prvi (kljuc)
    graph[first].append(second)
  return graph, first, second # vrati graf, prvi i drugi
```

```python
# eulerov ciklus -> balansiran i potpuno povezan
def findEulerianPath(graph):
  graph, first, second = RebalanceGraphFromEulerianPathToCycle(graph)
  eulerCycle = eulerianCycle(graph) # napravi eulerov krug od dobivenog grafa
  path = eulerCycle[0:-1]
  parts = eulerCycle.split(f"{str(first)}->{str(second)}", maxsplit=1)
  if parts[0] == '':
    path = parts[1]
  elif parts[1] == '':
    path = parts[0]
  else:
    first = str(first)
    second = str(second)
    part1 = parts[0] + first
    part1_trim = part1.split('->', maxsplit = 1)[1]
    part2 = second + parts[1]
    path = part2 + '->' + part1_trim
  return path
```

## [BA3I](https://rosalind.info/problems/ba3i/) - Find a k-Universal Circular String

```python
def getBinary(k):
  kmers = []
  for i in range(2**k):
    kmer = str(bin(i))[2:] # uzmi binarni zapis broja
    if len(kmer) < k:
      kmer = '0'*(k-len(kmer))+kmer # nadodaj nule na pocetak koje fale
    kmers.append(kmer)
  return kmers
```

```python
def kUniversalString(k):
  kmers = getBinary(k) #dohvati sve binarne zapise
  graph = constructGraph(kmers) # napravi debrujinov graf
  graph = sortDict(graph) # sortiaj
  cycle = eulerianCycle(graph) # napravi eulerov ciklus (BA3f)
  cycle = cycle.split('->')
  cycle = cycle[:len(cycle)-k+1] # uzmi do - (k-1) elementa
  res = cycle[0][:-1]
  for el in cycle:
    res += el[-1]
  return res
k = 4
kUniversalString(4)
```

## [BA3J](https://rosalind.info/problems/ba3j/) - Reconstruct a String from its Paired Composition

```python
def pairedGraph(pairs):
  D= defaultdict(list)
  for pair in pairs:
    a1, a2 = pair.split('|')
    p1 = tuple([a1[:-1], a2[:-1]]) # napravi tuple od prefiksa 1. i 2. elementa
    p2 = tuple([a1[1:], a2[1:]]) # napravi tuple od sufiksa 1. i 2. elementa
    D[p1].append(p2) # dodaj u rjecnik
  return D
```

```python
def stringFromPairedComposition(paired, k, d):
  path = findEulerianPath(paired) # pronadi eulerov put (BA3G)
  path = path.split('->') # napravu niz od puta
  path = [ast.literal_eval(x) for x in path] # stringove pretvori u tuples
  a1 = path[0][0][:-1]
  a2 = path[0][1][:-1]
  a1 += ''.join([x[0][-1] for x in path])
  a2 += ''.join([x[1][-1] for x in path])
  return a1+a2[-(k+d):]

k = 4
d = 2
pairs = [
'GAGA|TTGA',
'TCGT|GATG',
'CGTG|ATGT',
'TGGT|TGAG',
'GTGA|TGTT',
'GTGG|GTGA',
'TGAG|GTTG',
'GGTC|GAGA',
'GTCG|AGAT']
stringFromPairedComposition(pairedGraph(pairs),k, d) == 'GTGGTCGTGAGATGTTGA'
```

## [BA3K](https://rosalind.info/problems/ba3k/) - Generate Contigs from a Collection of Reads

```python
def constructDeBruijn(kmers):
  result = {} # rijecnik za spremanje rezultata
  for kmer in kmers: # za svaki kmer u listi kmera
    if kmer[:-1] not in result: # ako prefix nije u rezultatima
      # postavi kao kljuc prefiks, a kao vrijednost sufiks
      result[kmer[:-1]] = [kmer[1:]]
    else: # inace
      result[kmer[:-1]].append(kmer[1:]) # na postojeci prefiks nadodaj sufiks
  return result
```

```python
def remove_edge(adj_list, from_node, to_node):
  # u rijecniku ukloni pripadajucu vrijednost odredenog kljuca
  adj_list[from_node].remove(to_node)
  if not adj_list[from_node]: # ako je vrijednost za kljuc prazna
    del adj_list[from_node] # ukloni kljuc
  return adj_list # vrati listu
```

```python
def findInOutDegrees(adj_list):
  in_out_degrees = {}
  # za svaki par iz rijecnika kljuc - vijednost
  for source, targets in adj_list.items():
    # ako kljuc nije u novom rijecniku
    if source not in in_out_degrees:
      # postavi kljuc kao kljuc u novom rijecniku
      # i za vrijednost dodaj niz s elementima 0 i
      # brojem pripadajucih vrijednosti iz pocetnog rijecnika
      in_out_degrees[source] = [0, len(targets)]
    else: # inace
      # ako se pojavi ponovno isti kljuc, nadodaj jos na
      # drugi element i njegove vrijednosti
      in_out_degrees[source][1] += len(targets)
    # za svaku vrijednost iz pocetnog rijecnika za odredeni kljuc
    for target in targets:
      if target not in in_out_degrees: # ako vrijednosti nema medu kljucevima
        # dodaj ga u rijecnik s nizom kao vrijednosti 1, 0
        in_out_degrees[target] = [1, 0]
      else: # inace ako se vec nalazi u rijecniku nadodaj na prvi el. niza +1
        in_out_degrees[target][0] += 1
  return in_out_degrees
```

```python
# pronadi sve djelove koji se ne granaju
def findAllNonBranchingParts(adj_list):
  paths = []
  in_out_degrees = findInOutDegrees(adj_list)
  for v in list(in_out_degrees): # za sve jedinstvene vrijednosti
    # ako je vrijednost za kljuc razlicita od [1,1]
    if in_out_degrees[v] != [1, 1]:
      # i ako je drugi element iz vrijednosti veci od 0
      if in_out_degrees[v][1] > 0:
        while v in adj_list: # dok je jedinstvena vrijednost u grafu
          w = adj_list[v][0] # uzmi prvi element iz vrijednosti
          # put koji se ne grana je par kljuca i vrijednosti
          non_branching_path = [v, w]
          # izbaci iz grafa vrijednost i ako kljuc nema vise vrijednosti
          # izbaci vrijednost
          adj_list = remove_edge(adj_list, v, w)
          while in_out_degrees[w] == [1, 1]: # dok je vrijednost za kljuc [1,1]
            u = adj_list[w][0] # uzmi prvu vrijednost za taj kljuc
            # dodaj u ne granajuci put tu vrijednost
            non_branching_path.append(u)
            # ukloni iz grafa kombinaciju ovog kljuca i vrijednosti
            adj_list = remove_edge(adj_list, w, u)
            w = u # sada je novi kljuc vrijednost iz ovog koraka
          # na kraju dodaj negranajuci put u put
          paths.append(non_branching_path)
  return paths, adj_list # vrati put
```

```python
def maximalNonBranchingPaths(adj_list):
  paths, adj_list = findAllNonBranchingParts(adj_list)
  while adj_list: # dok graf ima elemenata
    # uzmi sve kljuceve i stavi ih u listu te uzmi prvi element
    start_node = list(adj_list)[0]
    current_node = adj_list[start_node][0] # uzmi njegovu vrijednost
    # ukloni ih iz grafa
    adj_list = remove_edge(adj_list, start_node, current_node)
    cycle = [start_node, current_node] # napravi listu od para
    while current_node != start_node: # dok je vrijednost razlicita od kljuca
      # uzmi prvu vrijednost koja se poklapa s vrijednoscu iz prethodnog koraka
      # vrijednost iz prethodnog koraka je sada kljuc
      target_node = adj_list[current_node][0]
      cycle.append(target_node) # nadodaj u listu i ovu vrijednost
      # ukloni ovu vrijednost i po potrebi odgovarajuci kljuc
      # (vrijednost iz prethodnog koraka)
      adj_list = remove_edge(adj_list, current_node, target_node)
      current_node = target_node # postavi da je novi kljuc sadasnja vrijednost
    paths.append(cycle) # u put nadodaj dobiveni niz
  return paths # vrati put
```

```python
def GenerateContigs(kmers):
  # za sve prefikse nadi odgovarajuce sufikse
  adj_list = constructDeBruijn(kmers) # kontruiraj debrujin graf
  paths = maximalNonBranchingPaths(adj_list) # pronadi najduzi negranajuci put
  contigs = []
  for path in paths: # za svaki put u putevima
    contig = path[0] # uzmi nulti element
    for edge in path[1:]: # za svaki element (osim nultog) u putu
      contig += edge[-1] # nadodaj njegovo zadnje slovo
    contigs.append(contig) # dodaj contig u konacni niz
  return contigs

def returnContigs(kmers):
  contigs = GenerateContigs(kmers) # generiraj contige
  contigs.sort() # sortiraj ih
  return " ".join(contigs) # vrati ispis u obliku stringa

cont = returnContigs(['AGA','ATG','ATG','TGT','TGG','CAT','GGA','GAT'])
print(cont)
```

```python
with open("rosalind_ba3k.txt", "r") as f:
  text = [line.strip() for line in f]

res = returnContigs(text)
res
```

## [BA3L](https://rosalind.info/problems/ba3l/) - Construct a String Spelled by a Gapped Genome Path

```python
def splitGappedPatterns(gappedPatterns): # razdvoji po | izmedu para
  splited = []
  for pair in gappedPatterns:
    splited.append(pair.split('|'))
  return splited
```

```python
def StringSpelledByGappedGenomePatterns(gappedPatterns, k, d):
  prefix = suffix = '' # prefiks i sufiks su prazan string
  # prolazimo kroz cijeli gappedPatterns kako bi kontruirali prefiks i sufiks
  # pozicija i vrijednost u listi - iteriraj
  for i, patterns in enumerate(gappedPatterns):
    if i != len(gappedPatterns)-1: # ako je i razlicit od zadnje pozicije
      prefix += patterns[0][0] # uzmi prvo slovo prvog elementa iz parova liste
      suffix += patterns[1][0] # uzmi prvo slovo drugog elementa iz parova liste
    else: # inace, ako je zadnja pozicija
      prefix += patterns[0] # u prefiks dodaj cijeli prvi element iz para
      suffix += patterns[1] # u sufiks dodaj cijeli drugi element iz para
  # provjera jesu li preklapanja savrsena ili ne
  # uzimamo d zadnjih od pozicije d u prefiksu
  for i in range(k+d, len(prefix)): # za od k+d pozicije do kraja prefiksa
    # ako slova na odredenim pozicijama nisu jednaka
    # i-ta pozicija u prefiksu je jednaka i-k-d poziciji u sufiksu
    if prefix[i] != suffix[i-k-d]:
      # ne moze se konstruirati string
      return "String cannot be constructed."
  # inace povezi prefiks i sufiks od znaka do kojeg se preklapaju
  return prefix + suffix[len(suffix)-k-d:]
```
