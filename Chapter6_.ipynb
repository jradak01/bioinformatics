## [BA6A](https://rosalind.info/problems/ba6a/) - Implement GreedySorting to Sort a Permutation by Reversals

```python
def format_num(x):
  # x je svaki element
  if x >= 0: # ako je x veci ili jednak nula
    return f"+{x}" # dodaj plus ispred broja
  else: # inace
    return f"{x}" # ostavi broj kakav je
```

```python
def rosalind_print_perm(permutations):
  strings = [] # string je prazan niz
  for perm in permutations: # za svaku permutaciju u permutacijama (nizovi)
    # nadodaj na niz obicne zagrade i stavi razmake izmedu elemenata niza
    # formatiraj element svaki pomocu funkcije f
    strings.append("(" + " ".join([format_num(x) for x in perm]) + ")")
  # vrati sve nizove u obliku stringa gdje je svaka  permutacija odvojena novim redom
  return "\n".join(strings)
```

```python
# ova funckija nam ispisuje step by step kako se radila zamjena
def greedy_sorting(str_permutation):
  # izvadi brojeve iz zagrada, rijesi se zagrada te razdvoji i pretvori u niz
  helper = [int(x) for x in str_permutation[1:-1].split()]
  S = []
  for i in range(0, len(helper)): # za svaki element u nizu
    # i == i+1 zato sto ocekujemo na 0-toj poziciji 1
    if helper[i] == i + 1: # ako je element i+1 niza na i-toj poziciji
      continue # preskoci i nastavi
    idx = i # spremi u varijablu idx trenutnu pozicju
    while True: # dok je istina
      # ako index sadrzi pravi element ili ako je na pozicji pravi element s neg
      # predznakom
      if helper[idx] == i + 1 or helper[idx] == -1 * (i + 1):
          break # izadi iz petlje
      idx += 1 # povecaj index
    # uzmi od trenutne pozicije do elementa na kojem se nalazi njegov par
    # (onaj koji je jednak poziciji) i cijeli taj komad zaokreni te
    # promijeni predznake (zato je prvi element u listi -1)
    mid = [-1 * x for x in helper[i : (idx + 1)]][::-1]
    # spoji taj zaokrenuti dio na ostatak niza
    helper = helper[0:i] + mid + helper[(idx + 1) :]
    S.append(helper.copy()) # na S nadodaj kopiju helpera
    if helper[i] < 0: # ako je pozicija u helperu manja od 0
      helper[i] = abs(helper[i]) # uzmi apsolutnu vrijednost od i-tog elementa
      S.append(helper.copy()) # nadodaj na S kopiju s apsolutnim vrijednostima
  return S # vrati S
```

## [BA6B](https://rosalind.info/problems/ba6b/) - Compute the Number of Breakpoints in a Permutation

```python
def rosalind_input(text):
  # izvadi brojeve iz zagrada, rijesi se zagrada te razdvoji i pretvori u niz
  tmp = [int(x) for x in text[1:-1].split(" ")]
  # dodaj nulu na temp i najveci element
  return [0] + tmp + [len(tmp) + 1]
```

```python
def is_adjacency(pair):
  # ako je razlika dobivenog para jednaka 1
  if pair[1] - pair[0] == 1:
      return True # vrati istina
  return False # inace vrati laz
```

```python
def n_adjacencies(permutation):
  count = 0 # postavi brojac na 0
  for i in range(len(permutation) - 1): # iteriraj po nizu
    # nadodaj +1 u brojac ako je je razlika para 1
    count += is_adjacency((permutation[i], permutation[i + 1]))
  return count # vrati brojac
```

```python
def n_breakpoints(permutation):
  # permutacije sadrze 0 i n+1 jer smo nadodali zato trebamo -1
  # vrati od boja elemenata u nizu oduzmi 1 i oduzmi broj permutacija
  return len(permutation) - 1 - n_adjacencies(permutation)

ans = n_breakpoints(
    rosalind_input('(+3 +4 +5 -12 -8 -7 -6 +1 +2 +10 +9 -11 +13 +14)'))
print(ans)
```

```python
with open("rosalind_ba6b.txt", "r", encoding='utf-8') as f:
  inlines = [x.strip("\n") for x in f.readlines()]
res = n_breakpoints(rosalind_input(inlines[0]))
res
```

## [BA6C](https://rosalind.info/problems/ba6c/) - Compute the 2-Break Distance Between a Pair of Genomes

```python
def parse_genome_graph(s):
  g = defaultdict(list)
  # razdvaja komponente na temelju zatvorene zagrade
  # svaka pronadena komponenta se pretvara u listu brojeva pomocu
  # component.split() i map(int, ...)
  # komponenta je oblika "1 2 3" i pretvara se u listu [1, 2, 3]
  for component in re.findall(r"\((.+?)\)", s):
    component = list(map(int, component.split()))
    for i in range(len(component) - 1):
      # dodavanje suprotnog brida u graf - svi susjedi povezani
      g[component[i]] += [-component[i + 1]]
      g[-component[i + 1]] += [component[i]]
    # zadnjem cvoru dodaj nulti - ciklicnost
    g[component[-1]] += [-component[0]]
    g[-component[0]] += [component[-1]]
  return g
```

```python
def find_component(node, graph, with_neg = False):
  q = [node] # dodaj pocetni cvor u listu za pretrazivanje
  if with_neg: visited = list()
  else: visited = set() # skup za spremanje posjecenih
  while q: # dok postoje elementi u q
    node = q.pop(0) # izbaci prvog iz liste
    if with_neg: visited.append(node)
    else: visited.add(node) # dodaj cvor u skup posjecenih
    for n in graph[node]: # dok postoje susjedni cvorovi cvora
      if with_neg: n = -n
      if n not in visited: # ako cvor nije u posjecenima
        q += [n] # dodaj cvor u listu za pretrazivanje
  return visited # vrati sve posjecene cvorove
```

```python
# spaja grafove p i q u jedan graficki prikaz razlomnih tocaka
# pretpostavlja se da oba grafa imaju iste cvorove
def breakpoint_graph(p, q):
  bg = {}
  for node in p.keys():
    bg[node] = p[node] + q[node]
  # vraca rezultirajuci graf bg koji predstavlja spojene razlomne tocke
  return bg
```

```python
# za izracunavanje udaljenosti dva prekida izmedu genoma
# udaljenost se mjeri brojem blokova koji su preuredeni kako bi se preslo s
# jednog genoma na drugi, petlja prolazi kroz sve komponente u grafu razlomnih
# tocaka, a svaka komponenta se broji kao zasebna komponenta
# Konacni rezultat je razlika izmeu broja blokova i broja komponenti
def two_break_distance(genomes):
  bg = breakpoint_graph(*genomes)
  nodes = set(bg.keys()) # skup cvorova u grafu razlomnih tocaka
  # broj cvorova - zbog 2 predznaka
  n_blocks = len(nodes) // 2
  n_components = 0 # broj komponenti
  while len(nodes): # dok ima cvorova
    # za svaki cvor pronalazi cvorove do kojih se moze doci iz trenutnog
    res = find_component(next(iter(nodes)), bg)
    # uklanjaju se cvorovi koji su dio pronadene komponente iz skupa nodes
    nodes = nodes - res
    # brojac n_components se povecava za 1 jer je pronadena nova komponenta
    n_components += 1
  # Vraca se razlika izmedu broja blokova (n_blocks) i broja pronadenih
  # komponenti (n_components) - ovo predstavlja udaljenost dva prekida
  # (two-break distance) izmedu genoma
  return n_blocks - n_components

inpt= ['(+1 +2 +3 +4 +5 +6)',
'(+1 -3 -6 -5)(+2 -4)']
genomes = [parse_genome_graph(s) for s in inpt]
two_break_distance(genomes)
```

```python
with open("rosalind_ba6c.txt", "r", encoding='utf-8') as f:
  inlines = [x.strip("\n") for x in f.readlines()]
genomes = [parse_genome_graph(s) for s in inlines]
res = two_break_distance(genomes)
res
```

## [BA6D](https://rosalind.info/problems/ba6d/) - Find a Shortest Transformation of One Genome into Another by 2-Breaks

```python
# uzima listu perm kao ulazni argument i vraca formatiranu verziju te liste
# input:  [1, -2, 3, 0]
# output (+1 -2 +3 +0)
def format_perm(perm):
  # f"{x:+}" - predznak
  return "(" + " ".join([f"{x:+}" for x in perm]) + ")"
```

```python
def find_connected_components(graph):
  nodes = set(graph.keys()) # skup cvorova u grafu
  while nodes: # dok ima cvorova
    # pronadi komponente za svaki cvor
    res = find_component(next(iter(nodes)), graph) #BA6C
    nodes = nodes - res # ukloni iz cvorova za pretragu one koji su istrazeno
    # vrati povezane komponente u svakom koraku
    yield res
```

```python
# nebanalni ciklusi - jednostavni ciklusi - barem 3 cvora i ne prolazi istim
# bridom vise od jednom
def non_trivial_cycle_nodes(graph):
  # za svaku povezanu komponentu
  for c in find_connected_components(graph): # BA6C
    # ako je ciklus nebanalan
    if len(c) > 2:
      # vrati povezanu komponentu
      return list(c)
  # nema nebanalinh ciklusa
  return None
```

```python
def format_genome_graph(g):
  nodes = set(g.keys()) # skup cvorova
  components = [] # komponente
  # izvrsava se sve dok ima cvorova u skupu nodes
  while nodes:
    # uzmi sve komp koje krecu od cvora
    comp = find_component(next(iter(nodes)), g, with_neg = True)
    nodes = nodes - set(comp) # smanji cvorove za pretragu s prijedenim
    nodes = nodes - set(-x for x in comp) # ukloni one i suprotnog predznaka
    components += [comp] # u komponente dodaj novopronadenu komponentnu
  # formatiraj svaku komponentu za ispis
  x = [format_perm(c) for c in components]
  return "".join(x)
```

```python
def add_edge(g, i, j): # dodavanje bridova izmedu cvorova
  g[i] += [j]
  g[j] += [i]
def del_edge(g, i, j): # uklanjanje bridova izmedu cvorova
  g[i].remove(j)
  g[j].remove(i)
```

```python
# koristi se za izracunavanje najkraceg transformacijskog puta izmedu dvaju
# genoma P i Q - petlja se izvrsava sve dok postoji nebanalni ciklus u grafu
# a svaki put kad se izmijene bridovi u grafu P - generira se formatirana
# verzija trenutnog grafa
def shortest_transform(P, Q):
  # napravi graf razlomnih tocaka iz permutacija P i Q
  bg = breakpoint_graph(P, Q) # BA6C
  # pronadi sve netrivijalne cikluse u grafu razlomnih tocaka
  nodes = non_trivial_cycle_nodes(bg)
  # vrati perm P kao prvi korak
  yield format_genome_graph(P) # (+1 -2 -3 +4)
  #  izvrsava sve dok ima cvorova u varijabli nodes
  while nodes:
    # j se postavlja na prvi cvor u nodes - ovaj cvor predstavlja proizvoljni
    # brid iz q bridova u nebanalnom ciklusu
    j = nodes[0]
    # i2 se postavlja na prvi cvor u Q koji je povezan s cvorom nodes[0]
    # ovaj cvor predstavlja brid iz referentnih bridova koji je povezan
    # s cvorom j
    i2 = Q[nodes[0]][0]
    # postavlja se na prvi cvor u P koji je povezan s cvorom j
    # cvor predstavlja brid iz referentnih bridova koji je povezan s cvorom j
    i = P[j][0]
    # j2 se postavlja na prvi cvor u P koji je povezan s cvorom i2
    # cvor predstavlja brid iz referentnih bridova koji je povezan s cvorom i2
    j2 = P[i2][0]
    # ukloni postojece veze
    del_edge(P, i, j)
    del_edge(P, i2, j2)
    # dodaj nove veze - prespoji
    add_edge(P, j, i2)
    add_edge(P, j2, i)
    # vrati novo stanje
    yield format_genome_graph(P)
    # azuriraj stanje
    bg = breakpoint_graph(P, Q)
    nodes = non_trivial_cycle_nodes(bg)

P = '(+1 -2 -3 +4)'
Q = '(+1 +2 -4 -3)'
for g in shortest_transform(parse_genome_graph(P), parse_genome_graph(Q)):
  print(g)
```

```python
with open("rosalind_ba6d.txt", "r", encoding='utf-8') as f:
  P, Q = [x.strip("\n") for x in f.readlines()]
for g in shortest_transform(parse_genome_graph(P), parse_genome_graph(Q)):
  print(g)
```

## [BA6E](https://rosalind.info/problems/ba6e/) - Find All Shared k-mers of a Pair of Strings

```python
def reverseComplement(dna):
  complements = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
  return ''.join([complements.get(x) for x in dna])
```

```python
# pronalazak zajednickih "k-mera" izmedu dvije DNA sekvence
def SharedKmers(k, dna1, dna2):
  result = []
  kmer_positions = defaultdict(list)
  # prolazak kroz prvu sekvencu
  for i in range(len(dna1) - k + 1):
    kmer = dna1[i:i+k]
    kmer_positions[kmer].append(i)
    if reverseComplement(kmer) in kmer_positions.keys():
      kmer_positions[reverseComplement(kmer)].append(i)
  # prolazak kroz drugu sekvencu
  for j in range(len(dna2) - k + 1):
    kmer = dna2[j:j+k]
    # ako se kmer nalazi i u prvoj i u drugoj seq
    if kmer in kmer_positions.keys():
      for pos in kmer_positions[kmer]: # dohvati sve pozicije iz rjecnika (za 1.)
        result.append((pos, j)) # dodaj pozicije u rezultat
    elif reverseComplement(kmer) in kmer_positions.keys():
      for pos in kmer_positions[reverseComplement(kmer)]:
        result.append((pos, j))
  return '\n'.join(str(x) for x in result)

k = 3
seq1 = 'AAACTCATC'
seq2 = 'TTTCAAATC'
result = SharedKmers(k, seq1, seq2)
print(result)
```

```python
with open("rosalind_ba6e.txt", "r", encoding='utf-8') as f:
  k, s1, s2 = [x.strip("\n") for x in f.readlines()]
result = SharedKmers(int(k), s1, s2)
print(result)
```

## [BA6F](https://rosalind.info/problems/ba6f/) - Implement ChromosomeToCycle

```python
# pretvara kromosomski zapis u ciklicki zapis tako da svaki blok iz kromosoma
# pretvara u dva cvora u ciklickom zapisu - pozitivni blokovi se pretvaraju u
# dva cvora, jedan paran i jedan neparan, dok se negativni blokovi takoder
# pretvaraju u dva cvora, ali s promjenom znaka i redoslijeda brojeva - funkcija
# vraca listu cvorova ciklickog zapisa
def chromosome_to_cycle(chromosome, external = True):
  nodes = []
  # petlja prolazi kroz svaki blok (segment) u kromosomu
  for block in chromosome:
    if block > 0: # ako je blok veci od 0
      # dodaj p[i], p[i+1]
      nodes.append(2 * block - 1)
      nodes.append(2 * block)
    else: # inace
      # dodaj p[i+1], p[i]
      nodes.append(-2 * block)
      nodes.append(-2 * block - 1)
  if external:
    return nodes
  return '(' + ' '.join(str(x) for x in nodes) +')'

chrom = '(+1 -2 -3 +4)'
chromosome = chrom.replace("(", "").replace(")", "")
chromosome = [int(x) for x in chromosome.split()]
chromosome_to_cycle(chromosome)
```

## [BA6G](https://rosalind.info/problems/ba6g/) - Implement CycleToChromosome

```python
# pretvara ciklicki zapis u kromosomski zapis tako sto svaki par cvorova u
# ciklickom zapisu pretvara u jedan blok u kromosomu - ako je prvi cvor manji od
# drugog cvora, to predstavlja pozitivan blok, a ako je prvi cvor veci ili
# jednak drugom cvoru, to predstavlja negativni blok -
# funkcija vraca listu blokova kromosoma
def cycle_to_chromosome(nodes, external = False):
  chromosome = []
  for i in range(0, len(nodes), 2):
    if nodes[i] < nodes[i + 1]: # ako je p[i] < p[i+1] -> pozitivan predznak
      chromosome.append(nodes[i + 1] // 2)
    else: # ako je p[i] > p[i+1] -> negativan predznak
      chromosome.append(-nodes[i] // 2)
  if external:
    return chromosome
  return '(' + ' '.join([format_num(x) for x in chromosome]) + ')'

nodes = '(1 2 4 3 6 5 7 8)'
nodes = nodes.replace("(", "").replace(")", "")
nodes = [int(x) for x in nodes.split()]
cycle_to_chromosome(nodes)
```

## [BA6H](https://rosalind.info/problems/ba6h/) - Implement ColoredEdges

**Obojeni bridovi:**

- ulaz: (+1 -2 -3)(+4 +5 -6)

- pretvori u cikliki zapis: [1, 2, 4, 3, 6, 5][7, 8, 9, 10, 12, 11]

- obojeni rubovi: (2, 4), (3, 6), (5, 1), (8, 9), (10, 12), (11, 7)

```python
def colored_edges(P):
  # prazna lista edges koja ce sadrzavati obojene bridove
  edges = list()
  # petlja prolazi kroz svaki kromosom u listi P
  for chromosome in P:
    # pretvara kromosom u ciklicki zapis i rezultat se sprema u listu nodes
    nodes = chromosome_to_cycle(chromosome, external = True) # BA6F
    # petlja prolazi kroz indekse j u koracima od 2, pocevsi od 1, sto znaci da
    # ce se uzeti u obzir svaki drugi element u listi nodes
    for j in range(1, len(nodes), 2):
      # provjerava se je li indeks j razlicit od broja elemenata u listi nodes
      # umanjenog za 1 - ova provjera se vrsi kako bi se osiguralo da se
      # posljednji brid ne dodaje posebno jer je povezan s pocetnim cvorom
      if j != len(nodes) - 1:
        # ako uvjet iz prethodne linije nije ispunjen, dodaje se brid u listu
        # edges koji je oblikovan kao lista s dva cvora, nodes[j] i nodes[j + 1]
        edges.append((nodes[j], nodes[j + 1]))
      # inace, j je jednak broju elemenata u listi nodes umanjenom za 1
      # sto znaci da se radi o posljednjem bridu
      else:
        # dodaje se posljednji brid u listu edges koji je oblikovan kao lista s
        # dva cvora, nodes[j] i pocetnim cvorom nodes[0]
        edges.append((nodes[j], nodes[0]))
  # na kraju funkcija vraca listu edges koja sadrzi obojene bridove
  return ' '.join(str(x) for x in edges)

P = '(+1 -2 -3)(+4 +5 -6)'
P = P[1:-1] # uklanjamo vanjske zagrade
P = P.split(')(') # razdvajamo po unutarnjim zagradama
# pisemo zasebno elemente i pretvaramo u brojeve
for i in range(len(P)):
  P[i] = [int(x) for x in P[i].split(' ')]
result = colored_edges(P)
print(result)
```

```python
with open("rosalind_ba6h.txt", "r", encoding='utf-8') as f:
  inline = f.readline().strip()
P = inline[1:-1] # uklanjamo vanjske zagrade
P = P.split(')(') # razdvajamo po unutarnjim zagradama
# pisemo zasebno elemente i pretvaramo u brojeve
for i in range(len(P)):
  P[i] = [int(x) for x in P[i].split(' ')]
result = colored_edges(P)
print(result)
```

## [BA6I](https://rosalind.info/problems/ba6i/) - Implement GraphToGenome

```python
def formatInput(edges):
  edges = '[' + edges + ']'
  edges = ast.literal_eval(edges)
  return edges
```

```python
# pretvara graficki prikaz genoma u kromosomski zapis genoma - prvo se ciklusi u
# grafickom prikazu genoma identificiraju i pohranjuju u listu Cycles - zatim se
# svaki ciklus pretvara u kromosomski zapis koristeci funkciju
# cycle_to_chromosome, a rezultirajuci kromosomi se dodaju u listu P
# na kraju se lista P vraca kao rezultat
def graph_to_genome(genomeGraph):
  P = [] # inicijalizira praznu listu P koja ce sadrzavati kromosome genoma
  # inicijalizira praznu listu Cycles koja ce sadrzavati cikluse
  # u grafickom prikazu genoma
  cycles = []
  # inicijalizira praznu privremenu listu temp koja ce privremeno pohranjivati
  # cvorove tijekom iteracije
  temp = []
  # petlja prolazi kroz indekse i u rasponu od 0 do duljine GenomeGraph
  # duljine broja parova
  for i in range(len(genomeGraph)):
    # provjerava se je li trenutni indeks i jednak duljini GenomeGraph umanjenoj
    # za 1 - ako je, to znaci da smo dosli do posljednjeg elementa u GenomeGraph
    if i == len(genomeGraph) - 1:
      # dodaje trenutni element GenomeGraph[i] u privremenu listu temp
      temp += genomeGraph[i]
      # dodaje privremenu listu temp u listu Cycles koja predstavlja jedan
      # ciklus u grafickom prikazu genoma
      cycles.append(temp)
    # ako prethodni uvjet nije ispunjen, provjerava se je li drugi element u
    # trenutnom paru cvorova jednak prvom elementu u sljedecem paru cvorova
    # povecanom za 1 ili smanjenom za 1
    # ovo provjerava kontinuitet u ciklickom zapisu genoma
    elif (genomeGraph[i][1] == genomeGraph[i + 1][0] + 1
      or genomeGraph[i][1] == genomeGraph[i + 1][0] - 1):
      # ako je uvjet ispunjen, dodaje trenutni element GenomeGraph[i]
      # u privremenu listu temp
      temp += genomeGraph[i]
    # ako nijedan od prethodnih uvjeta nije ispunjen, to znaci da smo dosli do
    # kraja ciklusa u grafickom prikazu genoma
    else:
      # dodaje trenutni element GenomeGraph[i] u privremenu listu temp
      temp += genomeGraph[i]
      # dodaje privremenu listu temp u listu cycles koja predstavlja jedan
      # ciklus u grafickom prikazu genoma
      cycles.append(temp)
      # resetira privremenu listu temp na praznu listu kako bi se pripremila za
      # sljedeci ciklus
      temp = []
  # petlja prolazi kroz svaki ciklus u listi cycles
  for cycle in cycles:
    # pretvara ciklicki zapis u kromosomski zapis - ulazni parametar je ciklus
    # Cycle, ali se prvo provodi rotacija ciklusa tako da se posljednji element
    # premjesta na pocetak liste
    chromosome = cycle_to_chromosome([cycle[-1]] + cycle[:-1]) #BA6G
    # dodaje se kromosom u listu P
    P.append(chromosome)
  # na kraju funkcija vraca listu P koja sadrzi kromosome genoma
  return ''.join(P)
edges = '(2, 4), (3, 6), (5, 1), (7, 9), (10, 12), (11, 8)'
edges = formatInput(edges)
result = graph_to_genome(edges)
result
```

```python
with open("rosalind_ba6i.txt", "r", encoding='utf-8') as f:
  inline = f.readline().strip()
edges = formatInput(inline)
result = graph_to_genome(edges)
result
```
